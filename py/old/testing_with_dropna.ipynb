# Data Prep

```python
import pandas as pd

df = pd.read_csv("../data/data.csv")
goals = ['goal1', 'goal2', 'goal3', 'goal4', 'goal5', 'goal6', 'goal7', 'goal8', 'goal9', 'goal10', 'goal11', 'goal12', 'goal13', 'goal14', 'goal15', 'goal16', 'goal17']
countries = df['Country'].unique()
print(f"the dataset has {len(countries)} countries and {len(goals)} goals")
```

```python
goals_with_missing_values = [goal for goal in goals if df[goal].isnull().sum() > 0]
countries_with_missing_values = df[df[goals_with_missing_values].isnull().any(axis=1)]['Country'].unique()
print(f"only {len(countries_with_missing_values)} countries have missing values and only for the goals {goals_with_missing_values}")

print(f"there are {impossible} impossible models and {possible} possible models")
```

```python
print(f"from the total of {len(countries)} countries and {len(goals)} goals we should be able to make {len(countries)*len(goals)} models")
impossible, possible = 0, 0
for country in countries:
    for goal in goals:
        if pd.isnull(df.loc[df['Country'] == country, goal].values[0]):
            impossible += 1
        else:
            possible += 1
print(f"but we can't make {impossible} models because of the countries with missing values")
print(f"so we are going to focus on the {(len(countries)*len(goals)) - impossible} models we can do based on the dataset we have")
```

```python
def check_possibilities(df):
    possible, impossible = 0,0
    for country in countries:
        for goal in goals:
            if pd.isnull(df.loc[df['Country'] == country, goal].values[0]):
                impossible += 1
            else:
                possible += 1
    return impossible, possible

print(f"With the dataset with size {df.shape} we can make {check_possibilities(df)[1]} models and we can't make {check_possibilities(df)[0]} models")
cleaned_df = pd.DataFrame()
for country in countries:
    for goal in goals:
        cleaned_df = pd.concat([cleaned_df, df[country].dropna(subset=[goal])])
print(f"but if we remove the rows with missing values we can make {check_possibilities(cleaned_df)[1]} models and we can't make {check_possibilities(cleaned_df)[0]} models")
```

```python
import pandas as pd
df_melted = pd.melt(df, id_vars=['Country', 'year'], 
                    value_vars=[f'goal{i}' for i in range(1, 18)], 
                    var_name='goal_number', value_name='goal_value')
df_melted.rename(columns={'Country': 'country'}, inplace=True)
df_melted = df_melted[['country', 'year', 'goal_value', 'goal_number']]

print(f"dataset original: {df.shape}")
print("17 objetivos multiplicados por 4008 linhas (países * anos) dá 68136")
print("dataset derretido: ", df_melted.shape)
print("dataset derretido tem 68136 linhas, então está correto")
```

```python
df_non_null = df_melted.dropna(subset=['goal_value'])
print(f"dataset original: {df_melted.shape}")
print(f"dataset sem valores nulos: {df_non_null.shape}")
```
