```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense

df = pd.read_csv("../data/data.csv")
window_sizes = [5, 10, 15, 20, 21, 22]
window_loss = {size: [] for size in window_sizes}
df_country = df.sort_values(by=["Country","year"])
data = df_country["goal2"].values  # Convert to numpy array
for window_size in window_sizes:
    if len(data) <= window_size:
        continue
    X, y = [], []
    for i in range(len(data) - window_size):
        X.append(data[i:i + window_size])
        y.append(data[i + window_size]) 
    X, y = np.array(X), np.array(y)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
    model = Sequential()
    model.add(Dense(10, input_dim=window_size, activation='relu'))
    model.add(Dense(1))  # Output prediction
    model.compile(loss='mean_squared_error', optimizer='adam')
    model.fit(X_train, y_train, epochs=100, verbose=0)
    loss = model.evaluate(X_test, y_test, verbose=0)
    window_loss[window_size].append(loss)
for window_size, losses in window_loss.items():
    avg_loss = np.mean(losses) if losses else float('nan')  # Avoid errors if no data
    print(f"Window Size {window_size}, Test Loss = {avg_loss:.4f}")
print("For every country <3")
print("Deep learning with one hidden layer with 10 nodes and relu activation (100 epochs)")
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense

df = pd.read_csv("../data/data.csv")
df = df.sort_values(by=["Country", "year"])

window_size = 20
X_windowed = []
y_windowed = []

for country in df['Country'].unique():
    country_data = df[df['Country'] == country]
    if len(country_data) >= window_size + 1:  # 20 years for X and 1 year for y (2023)
        X_windowed.append(country_data[(country_data['year'] >= 2000) & (country_data['year'] < 2020)].iloc[:, 2:].values)
        y_windowed.append(country_data[country_data['year'] == 2023].iloc[:, 2:].values)

X_windowed = np.array(X_windowed)
y_windowed = np.array(y_windowed)

print(f"X_windowed shape: {X_windowed.shape}")
print(f"y_windowed shape: {y_windowed.shape}")

X_train, X_test, y_train, y_test = train_test_split(X_windowed, y_windowed, test_size=0.2, shuffle=False)

model = Sequential()
model.add(Dense(64, input_dim=X_train.shape[1]*X_train.shape[2], activation='relu'))  # Input layer
model.add(Dense(32, activation='relu'))  # Hidden layer
model.add(Dense(17))  # Output layer, one prediction for each goal
model.compile(loss='mean_squared_error', optimizer='adam')
print(X_train.shape[1],X_train.shape[2])
print(X_train.shape[1]*X_train.shape[2])
model.fit(X_train.reshape(X_train.shape[0], X_train.shape[1]*X_train.shape[2]), y_train, epochs=20, batch_size=32, verbose=0)
```
