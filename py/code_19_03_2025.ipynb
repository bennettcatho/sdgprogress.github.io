```python
file_path = Path("../data/data.csv")
df = pl.read_csv(file_path).to_pandas().replace()
```

```python
df.filter(
    pl.col("id") == "AFG"
)
```

```python
import numpy as np
import matplotlib.pyplot as plt

cols = df.columns[3:]
country_id = "BRA"


data = df.filter(
    pl.col("id") == country_id
)[cols]
years = np.arange(2000, 2024)

for col in cols:
    plt.plot(years, data[col])

plt.legend(cols, loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```
